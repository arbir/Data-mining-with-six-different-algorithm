```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import silhouette_score
```

```
datos = pd.read_csv('/content/drive/MyDrive/Dataset/data.csv',index_col=0)

pd.set_option('display.max_columns', 32) 
datos.head()
```

```
datos.describe()
```

```
datos = datos.drop(['Unnamed: 32'], axis = 1)
datos.info()
```

```
datos['diagnosis'] = datos['diagnosis'].map({'M':1,'B':0})

from sklearn import preprocessing
datos_escalados = pd.DataFrame(preprocessing.scale(datos.iloc[:,1:32]))
datos_escalados.columns = list(datos.iloc[:,1:32].columns)

datos_escalados.describe()
```

```
from sklearn.cluster import Birch
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter= 4000, random_state = 2)
Y = tsne.fit_transform(datos_escalados)

birch1 = Birch(n_clusters=2)
bY = birch1.fit_predict(datos_escalados)
bY = [0 if x==1 else 1 for x in bY]
```

```
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(12, 8))

ax1.scatter(Y[:,0],Y[:,1],  c=bY, cmap = "jet", edgecolor = "None", alpha=0.35)
ax1.set_title('Birch Cluster')

ax2.scatter(Y[:,0],Y[:,1],  c = datos['diagnosis'], cmap = "jet", edgecolor = "None", alpha=0.35)
ax2.set_title('Actual Cluster')

plt.show()
```

```
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(datos['diagnosis'], bY))
print(classification_report(datos['diagnosis'], bY))
```

```
birch1 = Birch(n_clusters=2, branching_factor=1000, threshold=0.8,)
bY = birch1.fit_predict(datos_escalados)
bY = [0 if x==1 else 1 for x in bY]
print(confusion_matrix(datos['diagnosis'], bY))
print(classification_report(datos['diagnosis'], bY))
```
